## 1 - Função hipótese linear

```python
# Função hipótese de uma reta através da uma regressão linear sobre estes pontos e também usando a reta proposta
from scipy.interpolate import interp1d
from scipy.optimize import curve_fit

# Transformando array numpy de Multidimencional para unidimencional
x_uni = x.flatten()
y_uni = y.flatten()

# Criando uma função de interpolação linear para aproximar destes valores de amostras
interp_func = interp1d(x_uni, y_uni, kind='linear', fill_value='extrapolate')

# Pontos para a interpolação, eixo horizontal, indo de -1 a 1 aproximadamente
x_interp = np.linspace(min(x_uni), max(x_uni), 100)

# Calcular os valores interpolados
y_interp = interp_func(x_interp)

# Função linear para ajuste
def linear_func(x, a, b):
    return a * x + b

# Agora a função hipótese
# Ajuste da curva e associação a variavel de parámetros para coeficiente angular e linear
params, covariance_linear = curve_fit(linear_func, x_uni, y_uni)

# Parâmetros ajustados a0 e a1, ou a e b
a_line, b_line = params

# Pontos para a reta ajustada
y_fit_linear = linear_func(x_interp, a_line, b_line)
# Reta proposta
y_proposto = linear_func(x_interp, -1, 2)

# Plotar os dados originais com a interpolação de todos pontos e a interpolação linear
plt.scatter(x, y, label='Dados originais', color='blue')
plt.plot(x_interp, y_interp, label='Interpolação Linear', color='red')
plt.plot(x_interp, y_fit_linear, label=f'Reta Ajustada: y = {a_line:.4f}x + {b_line:.4f}', color='black', linewidth=3, marker='o')
plt.plot(x_interp, y_proposto, label=f'Reta Proposta: y = -x + 2', color='green', linewidth=2, marker='_')
plt.plot()
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid()
plt.show()


```

## Analisando a covariância da reta

### Os elementos da diagonal principal (posições onde a linha é igual à coluna) fornecem a variância de cada parâmetro ajustado. Quanto menor a variância, maior a confiança no valor do parâmetro.

### Uma covariância positiva indica que os parâmetros tendem a variar juntos, enquanto uma covariância negativa indica que eles tendem a variar inversamente.

```python
# Printanto o valor de covariância
print('matriz de covariância:\n',covariance_linear)
# Analisando o grau de incerteza sobre os pontos
param_errors = np.sqrt(np.diag(covariance_linear))
print('\nparametros de erros:\n',param_errors)
```

```python
# Propondo uma função hipótese de uma exponencial sobre estes pontos
from scipy.optimize import curve_fit

# Função exponencial para ajuste
def exponential_func(x, a, b, c):
    return a * np.exp(b * x) + c

# Ajuste da curva
params, covariance_exp = curve_fit(exponential_func, x_uni, y_uni, maxfev=5000)

# Parâmetros ajustados
a, b, c = params

# Pontos para a função exponencial ajustada
x_fit_exp = np.linspace(min(x), max(x), 100)
y_fit_exp = exponential_func(x_fit_exp, a, b, c)

# Plotar os dados originais e a função exponencial ajustada
plt.scatter(x, y, label='Dados originais', color='blue')
plt.plot(x_interp, y_interp, label='Interpolação Linear', color='red')
plt.plot(x_interp, y_fit_linear, label=f'Reta: y = {a_line:.4f}x + {b_line:.4f}', color='black', linewidth=5)
plt.plot(x_interp, y_proposto, label=f'Reta Proposta: y = -x + 2', color='green', linewidth=2, marker='_')
plt.plot(x_fit_exp, y_fit_exp, label=f'Exp: y = {a:.2f} * e^{b:.4f}x + {c:.2f}', color='orange',linewidth=1)
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

### Minimizando o erro

A seguir, iremos minimizar o erro do nosso modelo de regressão, treinando-o com os dados fornecidos.

Para treinar nosso modelo de regressão linear, precisamos definir

+ (i) o modelo,
+ (ii) a função de erro e
+ (iii) o procedimento (i.e., algoritmo) de treinamento.

### Definindo o modelo de regressão linear

O modelo definido abaixo através da classe `Model` representa a função hipótese $\hat{y} = a_0 + a_1 x$.

Portanto, o modelo gera um valor de $\hat{y}$ para determinados $a_0$, $a_1$ e $x$ quando é chamado.

## Modelo linear

```python
#importanto tensorflow
import tensorflow as tf

# Definindo nosso palpite inicial conforme encontrado anteriormente pelas interpolações liner
A0_INICIAL = 10.0
A1_INICIAL = -10.0

# Define nosso modelo de regressão.
class Model(object):
    def __init__(self, a0=A0_INICIAL, a1=A1_INICIAL):
        # Inicializando os pesos como variáveis do Tensorflow.
        self.a0 = tf.Variable(a0)
        self.a1 = tf.Variable(a1)

    # O método call é chamado toda a vez que passamos um valor de x para o modelo.
    def __call__(self, x):
        # Implementação da função hipótese.
        return self.a0 + self.a1*x
```

## Modelo exponencial (AINDA NÃO COMPILADO E NÃO SERÁ USADO, POIS NÃO É UMA ESCOLHA BOA)

```python
A0_INICIAL = 3145.94
A1_INICIAL = -0.0003
A2_INICIAL = -3143.96

# Define modelo de regressão.
class Model_exp(object):
    def __init__(self, a0=A0_INICIAL, a1=A1_INICIAL, a2=A2_INICIAL):
        # Inicializando os pesos como variáveis do Tensorflow.
        self.a0 = tf.Variable(a0)
        self.a1 = tf.Variable(a1)
        self.a2 = tf.Variable(a2)

    # O método call é chamado toda a vez que passamos um valor de x para o modelo.
    def __call__(self, x):
        # Implementação da função hipótese.
        return self.a0 * np.exp(self.a1 * x) + self.a2
```

## Definindo a função de erro

```python
# Definindo função de erro.
def erro(predicted_y, target_y):
    return tf.reduce_mean(tf.square(predicted_y - target_y))
```

### Definindo o procedimento (algoritmo) de treinamento

Aqui, o algoritmo do gradiente descendente é usado para o treinamento do modelo.

Função chamada `train` que executa **uma iteração do gradiente descendente** usando **TODAS** as amostras em nosso conjunto de dados (ou seja, usa todos os valores $x$ e $y$ disponíveis para treinamento).

Uma iteração do algoritmo de otimização do gradiente descendente envolve:

1. o cálculo do erro
2. o cálculo das derivadas da função de erro em relação ao pesos individuais (i.e., $a_0$ e $a_1$)
3. a atualização dos pesos do modelo com as derivadas calculadas (i.e., vetor gradiente)

```python
# Definindo o processo de treinamento (ou atualização dos pesos)
def train(model, inputs, outputs, learning_rate):
    with tf.GradientTape() as t:
        # Erro para o atual par de pesos.
        current_error = erro(model(inputs), outputs)

        # Aqui é onde diferenciamos a função de erro em relação aos pesos do modelo.
        da0, da1 = t.gradient(current_error, [model.a0, model.a1])      # derro/da0, derro/da1

        # E aqui é onde atualizamos os pesos do modelo com base no passo de aprendizagem escolhido.
        model.a0.assign_sub(learning_rate * da0)   # model.a0 = model.a0 - learning_rate*da0
        model.a1.assign_sub(learning_rate * da1)   # model.a1 = model.a1 - learning_rate*da1

        # Retorna o erro atual.
        return current_error
```

### Treinando o modelo

Em seguida, instanciamos o modelo.

Na sequência, treinamos o modelo por várias épocas. Para cada época, calculamos e armazenamos os valores dos pesos ($a_0$ e $a_1$) e o erro.

```python
# Passo de aprendizagem.
LEARNING_RATE = 0.05

# Número de épocas.
epochs = 300

# Instancia o modelo.
model = Model()

# Coletando o histórico de valores a0 e a1 para plotar mais tarde.
list_a0, list_a1 = [], []
erros = []

# Laço (loop) de treinamento.
for epoch in range(epochs):

    # Armazena os valores dos pesos.
    list_a0.append(model.a0.numpy())
    list_a1.append(model.a1.numpy())

    # Iteração de treinamento.
    current_error = train(model, x, y, learning_rate=LEARNING_RATE)

    # Armzena os erros ao longo das épocas de treinamento.
    erros.append(current_error)

    # Imprime os erros ao longo das épocas de treinamento.
    print('Época %2d: a0=%1.2f a1=%1.2f, loss=%2.5f' % (epoch, list_a0[-1], list_a1[-1], current_error))
```

## Plotando superfície de erro com a0 e a1 variando

### Caso aumente o número de pontos de amostras para a0 e a1 na linha de código
 ## `a0_range = np.linspace(10, -6, 1000)`

 ## `a1_range = np.linspace(10, -10, 1000)`
### verá um valor mais aproximado tanto de erro quanto de a0 e a1 para o ponto de mínimo da superfície como foi mostrado anteriormente no treino anteriormente

```python
from mpl_toolkits.mplot3d import Axes3D  # Importe o submódulo Axes3D

# Definindo uma grade de valores para a0 e a1 e 1000 amostras, quanto mais melhor resultado aproximado do que tinhamos
a0_range = np.linspace(10, -6, 1000)
a1_range = np.linspace(10, -10, 1000)
a0_grid, a1_grid = np.meshgrid(a0_range, a1_range)

# Função para calcular o erro quadrático médio (EQM)
def calcular_erro(a0, a1):
    y_pred = a0 + a1 * x
    erro = ((y_pred - y) ** 2).mean()
    return erro

# Erro para cada combinação de a0 e a1 para array multidimensional
error_grid = np.zeros_like(a0_grid)
for i in range(len(a0_range)):
    for j in range(len(a1_range)):
        a0 = a0_grid[i, j]
        a1 = a1_grid[i, j]
        # Calcule o erro para os valores de a0 e a1
        error_grid[i, j] = calcular_erro(a0,a1)

# Mínimo de erro e valores de a0 e a1 neste ponto
min_erro = np.min(error_grid)
constante_idx = np.unravel_index(np.argmin(error_grid), error_grid.shape)
ponto_a0 = a0_range[constante_idx[1]]
ponto_a1 = a1_range[constante_idx[0]]

# Plote a superfície de erro
fig = plt.figure(figsize=(15, 10))
ax = fig.add_subplot(111, projection='3d')  # Crie um eixo 3D
surf = ax.plot_surface(a0_grid, a1_grid, error_grid, cmap='bone', antialiased=False)
fig.colorbar(surf, ax=ax, shrink=0.5, aspect=5)
ax.set_xlabel('a0')
ax.set_ylabel('a1')
ax.set_zlabel('Erro')
ax.set_title(f'Superfície de Erro \nMínimo erro={min_erro:.10f}\na0={ponto_a0:.10f}\na1={ponto_a1:.10f}')
plt.show()
```
